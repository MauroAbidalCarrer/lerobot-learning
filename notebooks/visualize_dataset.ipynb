```python
import numpy as np
from numpy import ndarray
import plotly.express as px
from lerobot.common.datasets.lerobot_dataset import LeRobotDataset
```

```python
REPO_ID = "lerobot/pusht"
dataset = LeRobotDataset(REPO_ID)
```

Display the first frame of the first episde.

```python
img:np.ndarray = (
    dataset[0]["observation.image"]
    .numpy()
    .swapaxes(0, 2)
)
# img.shape
px.imshow(img)
```

Display the first  episode.

```python
first_ep_data_subset = LeRobotDataset(REPO_ID, episodes=[0])
def get_image(sample: dict) -> ndarray:
    return (
        sample["observation.image"]
        .numpy()
        .swapaxes(0, 2)
    )
first_episode_video = np.stack([get_image(sample) for sample in first_ep_data_subset], axis=0)
first_episode_video.shape
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

fig, ax = plt.subplots()
im = ax.imshow(first_episode_video[0])

def update(frame):
    im.set_array(first_episode_video[frame])
    return [im]

ani = FuncAnimation(fig, update, frames=first_episode_video.shape[0], interval=50, blit=True)
plt.close(fig)  # Prevents duplicate static image

# Display inline in notebook
HTML(ani.to_jshtml())
```
